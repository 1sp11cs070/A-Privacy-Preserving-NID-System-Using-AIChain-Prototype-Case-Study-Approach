```
# -------------------------------
# 📦 Install Required Packages
# -------------------------------
!pip install flask pyngrok

# -------------------------------
# 📚 Imports
# -------------------------------
from flask import Flask, request, jsonify
import hashlib
import datetime
import random
from pyngrok import ngrok

# -------------------------------
# 🚀 Initialize App
# -------------------------------
app = Flask(__name__)

# -------------------------------
# 🔗 Simple Blockchain Structure
# -------------------------------
class Block:
    def __init__(self, index, data, previous_hash):
        self.index = index
        self.timestamp = str(datetime.datetime.now())
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        block_string = str(self.index) + self.timestamp + str(self.data) + self.previous_hash
        return hashlib.sha256(block_string.encode()).hexdigest()

class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        return Block(0, "Genesis Block", "0")

    def add_block(self, data):
        previous_block = self.chain[-1]
        new_block = Block(len(self.chain), data, previous_block.hash)
        self.chain.append(new_block)

blockchain = Blockchain()

# -------------------------------
# 🤖 AI Fraud Detection
# -------------------------------
def ai_fraud_detection(user_data):
    nid = user_data.get("nid", "")

    risk_score = len(nid) * 10

    if len(nid) < 5:
        return {"status": "Suspicious", "risk_score": risk_score}
    elif risk_score > 80:
        return {"status": "Fraud Detected", "risk_score": risk_score}
    else:
        return {"status": "Safe", "risk_score": risk_score}

# -------------------------------
# 👤 Register Identity
# -------------------------------
@app.route('/register', methods=['POST'])
def register():
    data = request.json

    if not data or "name" not in data or "nid" not in data:
        return jsonify({"error": "Invalid input"}), 400

    name = data["name"]
    national_id = data["nid"]

    identity_hash = hashlib.sha256((name + national_id).encode()).hexdigest()

    blockchain.add_block({
        "name": name,
        "nid_hash": identity_hash
    })

    return jsonify({
        "message": "Identity Registered Successfully",
        "hash": identity_hash
    })

# -------------------------------
# 🔍 Verify Identity
# -------------------------------
@app.route('/verify', methods=['POST'])
def verify():
    data = request.json

    name = data.get("name")
    national_id = data.get("nid")

    identity_hash = hashlib.sha256((name + national_id).encode()).hexdigest()

    for block in blockchain.chain:
        if isinstance(block.data, dict):
            if block.data.get("nid_hash") == identity_hash:
                ai_result = ai_fraud_detection(data)

                return jsonify({
                    "message": "Identity Verified",
                    "ai_analysis": ai_result
                })

    return jsonify({"message": "Identity Not Found"}), 404

# -------------------------------
# 📊 View Blockchain
# -------------------------------
@app.route('/chain', methods=['GET'])
def get_chain():
    chain_data = []

    for block in blockchain.chain:
        chain_data.append({
            "index": block.index,
            "timestamp": block.timestamp,
            "data": block.data,
            "hash": block.hash,
            "previous_hash": block.previous_hash
        })

    return jsonify(chain_data)

# -------------------------------
# 🌍 Start Server with Ngrok
# -------------------------------
public_url = ngrok.connect(5000)
print("🌐 Public URL:", public_url)

app.run()
```
